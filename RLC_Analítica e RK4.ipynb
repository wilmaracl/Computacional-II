# Introdução

Os circuitos que combinam Resistências (R), Capacitores (C) e Indutores (L) são denominados de circuitos RLC. O estudo destes circuitos é básico para o entendimento de muitos aparelhos da modernidade \cite{dorneles2008simulaccao}. Vamos primeiramente analisar o circuito RLC sem fonte e encontrar sua solução. Em seguida estudaremos o circuito RLC na montagem série com fonte de tensão constante. Vamos encontrar a sua solução e apresentar três exemplos. 

# O Circuito RLC sem fonte de tensão


Vamos analisar o circuito RLC, mas sem considerar a fonte de tensão. (Figura \ref{RLCserie}). 

\begin{figure}[H]	 \centering    \includegraphics[scale=1]  {RLC_serie.eps} \caption{Circuito RLC com fonte de tensão}  \label{RLCserie}  \end{figure}

A diferença de potencial nos três elementos deste circuito são:

\begin{equation} R: Ri(t) \\   L: L \frac{d i(t)}{dt}  \\   C: Vc  \\  \end{equation}


Partindo de um ponto qualquer do circuito e somando as diferenças de potenciais, ao retornar ao mesmo ponto a soma deve ser zero:
\begin{equation} 
Ri(t) + L\frac{d i(t)}{dt} + Vc = 0
\end{equation}

Considerando que
\begin{equation}   i(t) = C\frac{dVc}{dt} \end{equation} e
\begin{equation}   \frac{di(t)}{dt}=C\frac{d^{2}Vc}{dt^{2}}  \end{equation}

Obtemos uma equação diferencial na variável Vc(t): 

\begin{equation}  LC\frac{d^2Vc(t)}{dt^2} + RC\frac{d Vc(t)}{dt} + Vc(t) = 0  \end{equation}

dividindo por LC, teremos:
    
\begin{equation}  \frac{d^2Vc(t)}{dt^2} + \frac{R}{L}\frac{d Vc(t)}{dt} + \frac{1}{LC}Vc(t)  \end{equation}

Esta é uma Equação Diferencial Ordinária (EDO) de segunda ordem. Admitindo que i(t)=e^{st} sendo s uma variável complexa, possa ser uma solução para esta EDO, teremos:

\begin{equation} s^2e^s + \frac{R}{L}se^s + \frac{1}{LC}e^s = 0 \end{equation}

Obtemos a equação característica desta EDO:

\begin{equation} s^2 + \frac{R}{L}s + \frac{1}{LC} = 0 \end{equation}

Cujas raizes são:

\begin{equation} s1 = -\frac{R}{2L} + \sqrt[2]{(\frac{R}{2L})^2-\frac{1}{LC}} \end{equation}
e
\begin{equation} s2 = -\frac{R}{2L} + \sqrt[2]{(\frac{R}{2L})^2 - \frac{1}{LC}} \end{equation}

As soluções da EDO são:

\begin{equation} Vc(t)=e^{(-\frac{R}{2L} + \sqrt[2]{(\frac{R}{2L})^2-\frac{1}{LC}})*t} \end{equation}
e
\begin{equation} Vc(t)=e^{(-\frac{R}{2L} - \sqrt[2]{(\frac{R}{2L})^2-\frac{1}{LC}})*t} \end{equation}

Em função do valor do radicando podemos analisar três soluções diferentes:

## a) Amortecimento forte:

\begin{equation} 
(\frac{R}{2L})^2 > \frac{1}{LC}  \\
\end{equation}

Neste caso teremos dois expoentes reais e negativos, com as soluções: \\

\begin{equation} 
Vc1(t)=e^{(-\frac{R}{2L} + \sqrt[2]{(\frac{R}{2L})^2-\frac{1}{LC}})*t} \end{equation}
e  
\begin{equation} 
Vc2(t)=e^{(-\frac{R}{2L} - \sqrt[2]{(\frac{R}{2L})^2-\frac{1}{LC}})*t}
\end{equation}

A solução completa é a combinação linear das duas soluções: \\

\begin{equation} 
Vc(t) = k1*Vc1(t) + k2*Vc2(t)
\end{equation}


## b) Amortecimento fraco:

\begin{equation} 
(\frac{R}{2L})^2 < \frac{1}{LC}
\end{equation}

Neste caso teremos a raiz de um número negativo, produzindo os dois expoentes imaginários.

Façamos \begin{equation} \theta = \sqrt[2]{\frac{1}{LC} - (\frac{R}{2L})^2}  \end{equation}

Então teremos como solução a combinação linear:

\begin{equation} Vc(t)= k1*e^{(-\frac{R}{2L} + j\theta)t} + k2*e^{(-\frac{R}{2L} - j\theta)t}  \end{equation}

Considerando as identidades de Euler:
    
\begin{equation} e^{(-\frac{R}{2L} + j\theta t)}=e^{-(\frac{R}{2L})*t}*[cos(\theta t) + jsen(\theta t)]  \end{equation}
e
\begin{equation} e^{(-\frac{R}{2L} - j\theta)*t}=e^{-(\frac{R}{2L})*t}*[cos(\theta t) - jsen(\theta t)]  \end{equation}

Voltamos em Vc(t):
    
\begin{equation} Vc(t) = e^{-(\frac{R}{2L})*t}*[(k1+k2)cos(\theta t) - j(k1-k2)sen(\theta t)]  \end{equation}

Criando as constantes

\begin{equation} A=k1+k2  \end{equation}  e   \begin{equation}  B=j(k1-k2)  \end{equation}

finalmente temos:
    
\begin{equation}  Vc(t) = e^{-(\frac{R}{2L})*t}*[Acos(\theta t) + Bsen(\theta t)]  \end{equation}

## c) Amortecimento crítico:

\begin{equation}  (\frac{R}{2L})^2 = \frac{1}{LC}  \end{equation}

Neste caso a solução se torna:

\begin{equation}  Vc(t)=A*e^{(-\frac{R}{2L})} + B*te^{(-\frac{R}{2L})}  \end{equation}

# O circuito RLC série com fonte de tensão constante V

Analisamos agora o circuito RLC com uma fonte de tensão. Considerando a fonte com tensão constante de valor V, a equação do circuito se torna: 

\begin{equation} \frac{d^2Vc(t)}{dt^2} + \frac{R}{L}\frac{d Vc(t)}{dt} + \frac{1}{LC}Vc(t) =\frac{V}{LC}  \end{equation}

Esta é uma ED para a qual já conhecemos as soluções da EDO correspondente.

Vamos considerar as seguintes condições iniciais da diferença de potencial Vc no capacitor:

\begin{equation} Vc(0) = 0      e      \frac{d Vc(0)}{dt} = 0 \end{equation}

# Três exemplos de circuitos RLC série com tensão constante

## a) Exemplo de Amortecimento forte

Para este caso usaremos R = 4 Ohm, L = 1 H e C = (1/3) F. Façamos primeiro os cálculos:
    
\begin{equation}  -\frac{R}{2L} + \sqrt[2]{(\frac{R}{2L})^2-\frac{1}{LC}}  \end{equation}
fica:
\begin{equation}  4/(2*1) + \sqrt[2]{(\frac{4}{2*1})^2-\frac{1}{1*(1/3)}} = -2 + 1 = -1  \end{equation}

e também

\begin{equation}  -\frac{R}{2L} - \sqrt[2]{(\frac{R}{2L})^2-\frac{1}{LC}}    \end{equation}    
que resulta:
\begin{equation}  -2-1 = -3  \end{equation}

A diferença de potencial no capacitor Vc(t) será: 

\begin{equation}  Vc(t) = Ae^{-t} + Be^{-3t} + V  \end{equation}

Usando a condição incial    Vc(0) = 0 
 A + B + V = 0    temos   B = -V - A
    
Usando a condição inicial  \begin{equation}  \frac{d Vc(0)}{dt} = 0  \end{equation}

-A -3B = 0        temos    A = -3B

Resolvendo estas duas equações obtemos:
    
B = V/2  e    A = (-3/2)V 

A solução para a diferença de potencial no capacitor fica:
    
\begin{equation}  (Vc(t) = (-3/2)Ve^{-t} + (V/2)e^{-3t} +V  \end{equation}

O programa Python abaixo calcula e apresenta a tensão no capacitor obtida por solução literal (linha contínua) e por método numérico Runge-Kutta (sinais + ).


```python
# Runge Kutta de Quarta Ordem - Amortecimento Forte
# Equação original: d**2y/dx**2 + 4*dy/dx + 3y = 30; condições iniciais: y(0) = 0, dy/dx(0) = 0, h = 0.25
# Seja dy/dx = z
# então, d**2y/dx**2 = dz/dx
# dz/dx + 4z + 3y = 30
# quebre em duas funções: f1(x,y,z) = dy/dx = z; e f2(x,y,z) = dz/dx = 30 - 4*z - 3*y
import numpy, math, matplotlib
from pylab import *

# Definindo a função que integra o sistema de EDOs 
def SistRK4(f, n_edos, n, x0, xf, y0):
    import numpy as np
    X = np.linspace(x0,xf,n)
    Y = np.zeros((n,n_edos), dtype=float)
    h = X[1] - X[0]
    # vetores auxiliares
    k1y = np.zeros((n_edos,1), dtype=float) 
    k2y = np.zeros((n_edos,1), dtype=float)
    k3y = np.zeros((n_edos,1), dtype=float)
    k1  = np.zeros((n_edos,1), dtype=float) 
    k2  = np.zeros((n_edos,1), dtype=float)
    k3  = np.zeros((n_edos,1), dtype=float)
    k4  = np.zeros((n_edos,1), dtype=float)
        
    for j in np.arange(0,n_edos,1):
        Y[0,j] = y0[j]
        
    for i in np.arange(0,n-1,1):
        for j in np.arange(0,n_edos,1):
            k1[j]  = h * f(j,X[i],Y[i,:])
            k1y[j] = Y[i,j]+0.5*k1[j]
        for j in np.arange(0,n_edos,1): 
            k2[j]  = h * f(j,X[i]+0.5*h,k1y[:])
            k2y[j] = Y[i,j]+0.5*k2[j]
        for j in np.arange(0,n_edos,1):           
            k3[j]  = h * f(j,X[i]+0.5*h,k2y[:])
            k3y[j] = Y[i,j] + 0.5*k3[j]
        for j in np.arange(0,n_edos,1):            
            k4[j] = h * f(j,X[i]+h,k3y[:])
        for j in np.arange(0,n_edos,1):     
            Y[i+1,j]  = Y[i,j] +(k1[j]+2*(k2[j]+k3[j])+k4[j])/6.0
            
    return X,Y


def edo1(i,x,y):
    import math as math
    # Considere as duas equacoes
    # y" = - 4y' - 3y + 30
    # Faz-se y =y1, y1'= y2
    #        y2' =  - 4* y' - 3y + 30
    # No python os índices iniciema em zero, logo
    # y1 = y[0] e y2 = y[1]
    #print(y.shape,y[0],y[1])
    if i==0:
        R = y[1]             # y(2) = \dot{y1}.
    elif i==1:
        R = -4*y[1] - 3*y[0] + 30   
    return R

# Aqui inicia o programa principal:
%matplotlib inline
# Implementando o método de RK4 para um sistema de EDOs
import numpy as np
import math as math

# Vamos definir as condições iniciais
x0  = 0   # Valor inicial do intervalo 
xf  = 5  # Valor final do intervalo
y0  = 0   # Condição inicial y(x0) = y0 = y(0)
y10  = 0
nx  = 50     # Número de pontos
n_edos = 2     # Numero de EDOS de 1a ordem
# Criando vetor y0 com condições iniciais
Y0 = np.zeros((n_edos,1), dtype=float)
Y0[0] = y0
Y0[1] = y10

# Integrando SistRK4(f, m, n, x0, xf, y0):
t, xs = SistRK4(edo1, n_edos, nx, x0, xf, Y0)

# Início da construção do gráfico dos resultados
import pylab as plt
largura = 10  # Largura da figura
altura  = 6   # Altura da figura
plt.figure(figsize =(largura, altura))     # Define o tamanho do gráfico
plt.grid(b='on',axis='both')     # Habilita a grade 
plt.plot(t, xs[:,0], '+b', label='Amortecimento Forte usando RK4')



# Monta o gráfico da função analítica de Amortecimento Forte
v=10.0   # Define a tensão da fonte
x = np.linspace(0,5,num=50)   # Define o vetor x com todos os pontos a calcular a função
R=4.0 ;  L=1.0 ; C=1.0/3.0   # Valores de RLC para Amortecimento Forte
alfa = R/(2*L)  ;  teta = math.sqrt( (R/(2*L))**2 - 1/(L*C) )
ya = [(v/2) * (2-3*(np.exp((-alfa+teta)*t)) + np.exp((-alfa-teta)*t) ) for t in x]
plt.plot(x,ya,'r-')    
    
    
plt.xlabel('t') 
plt.ylabel('y')
plt.legend()
plt.show()
```

## b) Exemplo de amortecimento fraco

Para este caso usaremos R = 4 Ohm, L = 1 H e C = (1/8) F. Façamos primeiro os cálculos:
                   
\begin{equation}   \theta = \sqrt[2]{\frac{1}{LC} - (\frac{R}{2L})^2} \end{equation} fica:
\begin{equation}   \theta = \sqrt{\frac{1}{1*(1/8)} - (\frac{4}{2*1})^2} = \sqrt{8-4}   \end{equation}   
\begin{equation}  \theta = 2  \end{equation}

A diferença de potencial no capacitor Vc(t) é:

\begin{equation} Vc(t) =  e^{-(\frac{R}{2L})*t}*(Acos(\theta t) + Bsen(\theta t)) + V  \end{equation}

Usando a condição incial Vc(0) = 0 : 
\begin{equation} A + V = 0  \end{equation}temos  \begin{equation}  A = -V  \end{equation}

\begin{equation}  \frac{d Vc(0)}{dt} = 0    \end{equation}    
\begin{equation}  -(R/2L)A + 1B*\theta = 0  \end{equation}  
\begin{equation}  B = [(R/2L)/\theta]*A     \end{equation}   temos:

B = [2 / 2] * (-V)  e    B = - V

A solução para a diferença de potencial no capacitor fica:

\begin{equation}  Vc(t)=V-Ve^{-2t}*(cos(2t)+sen(2t))  \end{equation}

O programa abaixo calcula e apresenta a tensão no capacitor obtida por solução literal (linha contínua) e por método numérico Runge-Kutta (sinais + ).


```python
# Runge Kutta de Quarta Ordem - Amortecimento Fraco
# Equação original: d**2y/dx**2 + 4*dy/dx + 8y = 80; condições iniciais: y(0) = 0, dy/dx(0) = 0, h = 0.25
# Seja dy/dx = z
# então, d**2y/dx**2 = dz/dx
# dz/dx + 4z + 8y = 80
# quebre em duas funções: f1(x,y,z) = dy/dx = z; e f2(x,y,z) = dz/dx = 80 - 4*z - 8*y
import numpy, math, matplotlib
from pylab import *

# Definindo a função que integra o sistema de EDOs 
def SistRK4(f, n_edos, n, x0, xf, y0):
    import numpy as np
    X = np.linspace(x0,xf,n)
    Y = np.zeros((n,n_edos), dtype=float)
    h = X[1] - X[0]
    # vetores auxiliares
    k1y = np.zeros((n_edos,1), dtype=float) 
    k2y = np.zeros((n_edos,1), dtype=float)
    k3y = np.zeros((n_edos,1), dtype=float)
    k1  = np.zeros((n_edos,1), dtype=float) 
    k2  = np.zeros((n_edos,1), dtype=float)
    k3  = np.zeros((n_edos,1), dtype=float)
    k4  = np.zeros((n_edos,1), dtype=float)
        
    for j in np.arange(0,n_edos,1):
        Y[0,j] = y0[j]
        
    for i in np.arange(0,n-1,1):
        for j in np.arange(0,n_edos,1):
            k1[j]  = h * f(j,X[i],Y[i,:])
            k1y[j] = Y[i,j]+0.5*k1[j]
        for j in np.arange(0,n_edos,1): 
            k2[j]  = h * f(j,X[i]+0.5*h,k1y[:])
            k2y[j] = Y[i,j]+0.5*k2[j]
        for j in np.arange(0,n_edos,1):           
            k3[j]  = h * f(j,X[i]+0.5*h,k2y[:])
            k3y[j] = Y[i,j] + 0.5*k3[j]
        for j in np.arange(0,n_edos,1):            
            k4[j] = h * f(j,X[i]+h,k3y[:])
        for j in np.arange(0,n_edos,1):     
            Y[i+1,j]  = Y[i,j] +(k1[j]+2*(k2[j]+k3[j])+k4[j])/6.0
            
    return X,Y


def edo1(i,x,y):
    import math as math
    # Considere as duas equacoes
    # y" = - 4* y' - 8y + 80
    # Faz-se y =y1, y1'= y2
    #        y2' =  - 4y' - 8y + 80
    # No python os índices iniciema em zero, logo
    # y1 = y[0] e y2 = y[1]
    #print(y.shape,y[0],y[1])
    if i==0:
        R = y[1]             # y(2) = \dot{y1}.
    elif i==1:
        R = -4*y[1] - 1000*y[0] + 10000   
    return R

# Aqui inicia o programa principal:
%matplotlib inline
# Implementando o método de RK4 para um sistema de EDOs
import numpy as np
import math as math

# Vamos definir as condições iniciais
x0  = 0   # Valor inicial do intervalo 
xf  = 5   # Valor final do intervalo
y0  = 0   # Condição inicial y(x0) = y0 = y(0)
y10  = 0
nx  = 10000   # Número de pontos
n_edos = 2     # Numero de EDOS de 1a ordem
# Criando vetor y0 com condições iniciais
Y0 = np.zeros((n_edos,1), dtype=float)
Y0[0] = y0
Y0[1] = y10

# Integrando SistRK4(f, m, n, x0, xf, y0):
t, xs = SistRK4(edo1, n_edos, nx, x0, xf, Y0)

# Início da construção do gráfico dos resultados
import pylab as plt
largura = 10  # Largura da figura
altura  =  6  # Altura da figura
plt.figure(figsize =(largura, altura))     # Define o tamanho do gráfico
plt.grid(b='on',axis='both')     # Habilita a grade 

plt.plot(t, xs[:,0], '+r', label='Amortecimento Fraco usando RK4')


# Monta o gráfico da função analítica de Amortecimento Fraco
v=10.0   # Define a tensão da fonte
x = np.linspace(0,5,num=500)   # Define o vetor x com todos os pontos a calcular a função
R=4 ;  L=1 ; C=1/1000
alfa = R/(2*L)  ;  teta = math.sqrt( (1/(L*C)) - (alfa**2) )
y = [v - v * np.exp(-alfa*t) * ( math.cos(teta*t) + (alfa/teta)*math.sin(teta*t)) for t in x]
print (alfa,teta)
plt.plot(x,y,'b-')  

plt.xlabel('t') 
plt.ylabel('y')
plt.legend()
plt.show()
```

## c) Exemplo de amortecimento críco

Para este caso usaremos R = 4 Ohm, L = 1 H e C = (1/4) F. A equação da diferença de potencial no capacitor é:

\begin{equation}  Vc(t)=A*e^{(-\frac{R}{2L})} + B*te^{(-\frac{R}{2L})} + V  \end{equation}

Com os valores do exemplo, temos:
    
\begin{equation}  Vc(t)=A*e^{(-2t)} + B*te^{(-2t)} + V  \end{equation}

A + V = 0  temos  A = -V

\begin{equation} \frac{dVc(0)}{dt} = 0  \end{equation}

\begin{equation}  Vc(t)=V + 2Vte^{-2t} - Ve^{2t}  \end{equation}

O programa abaixo calcula e apresenta a tensão no capacitor obtida por solução literal (linha contínua) e pelo método numérico Runge-Kutta (sinais + ).


```python
# Runge Kutta de Quarta Ordem - Amortecimento Crítico
# Equação original: d**2y/dx**2 + 4*dy/dx + 4y = 40; condições iniciais: y(0) = 0, dy/dx(0) = 0, h = 0.25
# Seja dy/dx = z
# então, d**2y/dx**2 = dz/dx
# dz/dx + 4z + 4y = 40
# quebre em duas funções: f1(x,y,z) = dy/dx = z; e f2(x,y,z) = dz/dx = 40 - 4*z - 4*y
import numpy, math, matplotlib
from pylab import *

# Definindo a função que integra o sistema de EDOs 
def SistRK4(f, n_edos, n, x0, xf, y0):
    import numpy as np
    X = np.linspace(x0,xf,n)
    Y = np.zeros((n,n_edos), dtype=float)
    h = X[1] - X[0]
    # vetores auxiliares
    k1y = np.zeros((n_edos,1), dtype=float) 
    k2y = np.zeros((n_edos,1), dtype=float)
    k3y = np.zeros((n_edos,1), dtype=float)
    k1  = np.zeros((n_edos,1), dtype=float) 
    k2  = np.zeros((n_edos,1), dtype=float)
    k3  = np.zeros((n_edos,1), dtype=float)
    k4  = np.zeros((n_edos,1), dtype=float)
        
    for j in np.arange(0,n_edos,1):
        Y[0,j] = y0[j]
        
    for i in np.arange(0,n-1,1):
        for j in np.arange(0,n_edos,1):
            k1[j]  = h * f(j,X[i],Y[i,:])
            k1y[j] = Y[i,j]+0.5*k1[j]
        for j in np.arange(0,n_edos,1): 
            k2[j]  = h * f(j,X[i]+0.5*h,k1y[:])
            k2y[j] = Y[i,j]+0.5*k2[j]
        for j in np.arange(0,n_edos,1):           
            k3[j]  = h * f(j,X[i]+0.5*h,k2y[:])
            k3y[j] = Y[i,j] + 0.5*k3[j]
        for j in np.arange(0,n_edos,1):            
            k4[j] = h * f(j,X[i]+h,k3y[:])
        for j in np.arange(0,n_edos,1):     
            Y[i+1,j]  = Y[i,j] +(k1[j]+2*(k2[j]+k3[j])+k4[j])/6.0
            
    return X,Y


def edo1(i,x,y):
    import math as math
    # Considere as duas equacoes
    # y" = - 4y' - 4y + 40
    # Faz-se y =y1, y1'= y2
    #        y2' =  - 4y' - 4y + 30
    # No python os índices iniciema em zero, logo
    # y1 = y[0] e y2 = y[1]
    #print(y.shape,y[0],y[1])
    if i==0:
        R = y[1]             # y(2) = \dot{y1}.
    elif i==1:
        R = -4*y[1] - 4*y[0] + 40   
    return R

# Aqui inicia o programa principal:
%matplotlib inline
# Implementando o método de RK4 para um sistema de EDOs
import numpy as np
import math as math

# Vamos definir as condições iniciais
x0  = 0   # Valor inicial do intervalo 
xf  = 5   # Valor final do intervalo
y0  = 0   # Condição inicial y(x0) = y0 = y(0)
y10  = 0
nx  = 50   # Número de pontos
n_edos = 2     # Numero de EDOS de 1a ordem
# Criando vetor y0 com condições iniciais
Y0 = np.zeros((n_edos,1), dtype=float)
Y0[0] = y0
Y0[1] = y10

# Integrando SistRK4(f, m, n, x0, xf, y0):
t, xs = SistRK4(edo1, n_edos, nx, x0, xf, Y0)

# Início da construção do gráfico dos resultados
import pylab as plt
largura = 10  # Largura da figura
altura  =  6  # Altura da figura
plt.figure(figsize =(largura, altura))     # Define o tamanho do gráfico
plt.grid(b='on',axis='both')     # Habilita a grade 
plt.plot(t, xs[:,0], '+g', label='Amortecimento Crítico usando RK4 ')



# Monta o gráfico da função analítica de Amortecimento Crítico
v=10.0   # Define a tensão da fonte
x = np.linspace(0,5,num=50)   # Define o vetor x com todos os pontos a calcular a função
R=4.0 ;  L=1.0 ; C=1.0/4.0  # Valores de RLC para Amortecimento crítico
alfa = R/(2*L)
yc = [v * (1 - 2*t*np.exp(-alfa*t) - np.exp(-alfa*t) ) for t in x]
plt.plot(x,yc,'r-')  



plt.xlabel('t') 
plt.ylabel('y')
plt.legend()
plt.show()
```

Nota-se que no amortecimento forte não acontece ondulação. À medida que o amortecimento diminui, passa-se por um ponto crítico, a partir do qual o amortecimento oscila, antes de chegar à estabilidade. Este é o ponto do amortecimento crítico.
